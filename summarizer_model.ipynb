```
import numpy as np
import tensorflow as tf
import pickle
```

```
dataset = '/content/drive/My Drive/dataset_story.pkl'
```

```
stories = pickle.load(open(dataset, 'rb'))
```

```
stories[1:6]
```

```
for i in stories:
  i['highlight'] = '<START>_' + i['highlight'] + '_<END>'
  
```

```
stories[1:6]
```

```
all_stories_list = []
all_highlights_list = []

all_stories = set()
all_highlights = set()

for i in stories:
  all_stories_list.append(i['story'])
  for word in i['story'].split():
    if word not in all_stories:
      all_stories.add(word)
      
for i in stories:
  all_highlights_list.append(i['highlight'])
  for word in i['highlight'].split():
    if word not in all_highlights:
      all_highlights.add(word)
            
```

```
input_words = sorted(list(all_stories))
target_words = sorted(list(all_highlights))
num_encoder_tokens = len(all_stories)
num_decoder_tokens = len(all_highlights)
```

```
num_encoder_tokens
```

```
num_decoder_tokens
```

```
max_encoder_seq_length = max([len(txt) for txt in input_words])
max_decoder_seq_length = max([len(txt) for txt in target_words])
```

```
max_encoder_seq_length
```

```
max_decoder_seq_length
```

```
encoder_input_data = np.zeros(
    (len(all_stories_list), max_encoder_seq_length),
    dtype='float32')
decoder_input_data = np.zeros(
    (len(all_highlights_list), max_decoder_seq_length),
    dtype='float32')
decoder_target_data = np.zeros(
    (len(all_highlights_list), max_decoder_seq_length, num_decoder_tokens),
    dtype='float32')
```
